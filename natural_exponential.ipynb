# Natural Exponential

## The Natural Exponential Function

The natural exponential function is defined as:

$$
f(x) = e^x
$$

### Why is it called "natural"?

1. **Based on Euler’s number (*e*)**  
   The constant $ e \approx 2.718 $ is like π’s lesser-known but equally legendary cousin. It naturally arises when we study continuous growth—like compounding interest infinitely fast or zooming into infinitesimally small changes in calculus.

2. **Self-replicating derivative**  
   This function is one of a kind: its rate of change is equal to its own value. In calculus terms:

$$
   \frac{d}{dx} e^x = e^x
$$

   That elegant property makes $e^x $ a cornerstone in calculus, differential equations, and systems modeling natural growth or decay.

3. **Continuous growth or decay**  
   If you're modeling something that grows—like populations, investments—or decays—like radioactive material or cooling coffee—and that change is proportional to the current amount, chances are you're working with an equation involving $ e^x $.

4. **Inverse of the natural logarithm**  
   The function $ \ln(x) $, known as the natural logarithm, is the inverse of the natural exponential function $ e^x $. That means they "undo" each other:

$$
   \ln(e^x) = x \quad \text{and} \quad e^{\ln(x)} = x
$$

   This inverse relationship is fundamental to solving exponential equations and appears frequently in calculus, statistics, and mathematical modeling.


# Graphing Natural Logarithm

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x range
x = np.linspace(-2, 2, 400)
y = np.exp(x)

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='f(x) = $e^x$', color='teal', linewidth=2)

# Add labels and title
plt.title('Natural Exponential Function', fontsize=14)
plt.xlabel('x')
plt.ylabel('$e^x$')
plt.axhline(0, color='gray', linewidth=0.8, linestyle='--')
plt.axvline(0, color='gray', linewidth=0.8, linestyle='--')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()

plt.show()
```

## Exponential Decay Function: `y = np.exp(-x)`

The expression `y = np.exp(-x)` represents the **exponential decay function**.

### What’s happening:

- The `np.exp()` part calls the natural exponential function $ e^x $.
- The negative sign in the exponent transforms exponential **growth** into **decay**:

$$
  y = e^{-x}
$$

This function starts high at $ x = 0 $ (since $ e^0 = 1 $), and then decreases rapidly as $ x $ increases—approaching zero but never reaching it. It's smooth, continuous, and always positive.

### Common applications:

- Radioactive decay and carbon dating  
- Cooling of heated objects  
- Absorption and decay curves in physics or economics  
- Kernel functions in machine learning (e.g., RBF/Gaussian kernel)


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x range
x = np.linspace(0, 5, 400)
y = np.exp(-x)

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = $e^{-x}$', color='indigo', linewidth=2)

# Add labels and title
plt.title('Exponential Decay Function', fontsize=14)
plt.xlabel('x')
plt.ylabel('$e^{-x}$')
plt.axhline(0, color='gray', linewidth=0.8, linestyle='--')
plt.axvline(0, color='gray', linewidth=0.8, linestyle='--')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()

plt.show()
```
